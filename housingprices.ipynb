```python
from sklearn.preprocessing import StandardScaler
NotScalerInput=x.iloc[:,-2:]
ScalerInput=x.drop(['ocean_proximity','City'],axis='columns')
Scaler=StandardScaler()
columns=['Median_Income','Median_Age','Tot_Rooms','Tot_Bedrooms','Population','Households','Latitude','Longitude','Distance_to_coast','Distance_to_LA','Distance_to_SanDiego','Distance_to_SanJose','Distance_to_SanFrancisco']
x[columns]=Scaler.fit_transform(ScalerInput)

```

```python
x[columns]= pd.DataFrame(x[columns])
x[columns].rename(columns = {0:columns[0],
                              1:columns[1],
                              2:columns[2],
                              3:columns[3],
                              4:columns[4],
                              5:columns[5],
                              6:columns[6],
                              7:columns[7],
                              8:columns[8],
                              9:columns[9],
                              10:columns[10],
                              11:columns[11],
                              12:columns[12]
                              }, inplace = True)
```

```python
x
```

```python
# Encoding
from sklearn.preprocessing import LabelEncoder
Encoder=LabelEncoder()
x['ocean_proximity'].unique()
```

```python
x['ocean_proximity']=Encoder.fit_transform(x['ocean_proximity'])
x['ocean_proximity'].unique()
```

```python
x['City']=Encoder.fit_transform(x['City'])
x['City'].unique()
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# Model Building
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)
# y_pred.head()
y_pred
```
